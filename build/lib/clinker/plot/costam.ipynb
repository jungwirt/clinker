```python
# Grupowanie według 't_name' i zliczanie ilości wystąpień w każdym klastrze
cluster_counts = df.groupby(['t_name', '__mclCluster']).size().reset_index(name='count')

# Wybieranie klastra z największą liczbą wystąpień dla każdego t_name
max_counts_idx = cluster_counts.groupby('t_name')['count'].idxmax()
result = cluster_counts.loc[max_counts_idx]
```

```python
def delete_column(df):
    return df.drop(columns=['count'])

def rename_column(df):
    return df.rename(columns={'__mclCluster': 'ID_of_cluster_with_most_frequent_t_name'})

# Wywołanie funkcji łańcuchowo
result_2 = result_2.pipe(delete_column).pipe(rename_column)
```

```python
filtered_df = wyniki_zlaczone.groupby('__mclCluster').filter(lambda x: len(x) >= 9)
```

```python
# Obliczenie liczby unikalnych klastrów
num_clusters = filtered_df['__mclCluster'].nunique()

# Obliczenie liczby wystąpień każdej domeny w klastrach
domain_counts = filtered_df['t_name'].value_counts(normalize=True)

# Listy na częste, częste w niektórych i rzadkie domeny
common_domains = []
some_common_domains = []
rare_domains = []

# Podział domen na listy na podstawie ich częstości występowania w klastrach
for domain, freq in domain_counts.items():
    if freq >= 0.9 * num_clusters:
        common_domains.append(domain)
    elif 0.3 * num_clusters < freq < 0.9 * num_clusters:
        some_common_domains.append(domain)
    else:
        rare_domains.append(domain)

print("Częste domeny:")
print(common_domains)
print("\nCzęste w niektórych domeny:")
print(some_common_domains)
print("\nRzadkie domeny:")
print(rare_domains)
```
